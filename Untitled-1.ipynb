```python
import pandas as pd

# Load your crop dataset
crop_df = pd.read_csv("District_Wise_Area_Production_Yield_Value_Spice_Crops.csv")

# Clean column names
crop_df.columns = crop_df.columns.str.strip().str.lower().str.replace(" ", "_")

print("Columns after cleaning:", crop_df.columns.tolist())

```

```python
crop_df = crop_df[["district_name", "area", "production"]]
```

```python
crop_df["district_name"] = crop_df["district_name"].str.title().str.strip()
crop_df["area"] = pd.to_numeric(crop_df["area"], errors="coerce")
crop_df["production"] = pd.to_numeric(crop_df["production"], errors="coerce")
```

```python
crop_df["year"] = 2021   # or whichever year it represents
```

```python
crop_df.to_csv("crop_data_clean.csv", index=False)
print("✅ Cleaned crop data saved as crop_data_clean.csv")
```

```python
import pandas as pd

# Load both datasets
crop_df = pd.read_csv("crop_data_clean.csv")
rain_df = pd.read_csv("rainfall_by_districts_2019.csv")

# Clean column names
crop_df.columns = crop_df.columns.str.strip().str.lower().str.replace(" ", "_")
rain_df.columns = rain_df.columns.str.strip().str.lower().str.replace(" ", "_")

# Rename for consistency
rain_df.rename(columns={"district": "district_name"}, inplace=True)

# Add year column to rainfall data (based on your description)
rain_df["year"] = 2017  # You can change this if needed

# Merge both datasets
merged_df = pd.merge(crop_df, rain_df, on=["district_name", "year"], how="inner")

# Save and show output
merged_df.to_csv("merged_crop_rainfall_data.csv", index=False)
print("✅ Merge successful! Merged file saved as 'merged_crop_rainfall_data.csv'")
print(merged_df.head())
```
